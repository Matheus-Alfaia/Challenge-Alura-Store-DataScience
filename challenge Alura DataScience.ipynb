### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# prompt: IA gere para mim o total de venda por categoria de cada uma das lojas e mostre o resultado de cada uma das lojas e gere um grafico de linha, com uma linha representado cada loja

import matplotlib.pyplot as plt

# Calcula o total de vendas por categoria para cada loja
vendas_loja1 = loja.groupby('Categoria do Produto')['Preço'].sum()
vendas_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
vendas_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
vendas_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()

# Plota o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(vendas_loja1.index, vendas_loja1.values, label='Loja 1')
plt.plot(vendas_loja2.index, vendas_loja2.values, label='Loja 2')
plt.plot(vendas_loja3.index, vendas_loja3.values, label='Loja 3')
plt.plot(vendas_loja4.index, vendas_loja4.values, label='Loja 4')

plt.xlabel('Categoria')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Categoria para cada Loja')
plt.legend()
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout() # Ajusta o layout para evitar sobreposição de elementos
plt.show()


```

# 3. Média de Avaliação das Lojas

```python
# prompt: faça um codigo para calcular a média de avalição de cada loja com o campo Avaliação da compra e gere um grafico de pontos

import pandas as pd
import matplotlib.pyplot as plt

# ... (código anterior permanece inalterado)

# 3. Média de Avaliação das Lojas

# Calcula a média de avaliação para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Cria um gráfico de pontos
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias_avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

plt.figure(figsize=(8, 6))
plt.scatter(lojas, medias_avaliacoes, s=50, c='skyblue', edgecolors='black')  # s define o tamanho dos pontos
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação das Compras por Loja')
plt.ylim(0, 5) # Define o limite do eixo y de 0 a 5 para avaliações

# Adiciona rótulos de dados nos pontos
for i, txt in enumerate(medias_avaliacoes):
    plt.annotate(f'{txt:.3f}', (lojas[i], medias_avaliacoes[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()

print(f"Média de Avaliação Loja 1: {media_avaliacao_loja1:.3f}")
print(f"Média de Avaliação Loja 2: {media_avaliacao_loja2:.3f}")
print(f"Média de Avaliação Loja 3: {media_avaliacao_loja3:.3f}")
print(f"Média de Avaliação Loja 4: {media_avaliacao_loja4:.3f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
x = 4
mais_vendido_loja1 = loja['Produto'].value_counts().head(x).reset_index()
mais_vendido_loja1.columns = ['Produto', 'Qtd']
mais_vendido_loja2 = loja2['Produto'].value_counts().head(x).reset_index()
mais_vendido_loja2.columns = ['Produto', 'Qtd']
mais_vendido_loja3 = loja3['Produto'].value_counts().head(x).reset_index()
mais_vendido_loja3.columns = ['Produto', 'Qtd']
mais_vendido_loja4 = loja4['Produto'].value_counts().head(x).reset_index()
mais_vendido_loja4.columns = ['Produto', 'Qtd']
# Asignar MultiIndex a las columnas para cada loja
mais_vendido_loja1.columns = pd.MultiIndex.from_product([['Loja 1'], mais_vendido_loja1.columns])
mais_vendido_loja2.columns = pd.MultiIndex.from_product([['Loja 2'], mais_vendido_loja2.columns])
mais_vendido_loja3.columns = pd.MultiIndex.from_product([['Loja 3'], mais_vendido_loja3.columns])
mais_vendido_loja4.columns = pd.MultiIndex.from_product([['Loja 4'], mais_vendido_loja4.columns])
# Concatenar lado a lado
mais_vendidos_juntas = pd.concat([mais_vendido_loja1, mais_vendido_loja2, mais_vendido_loja3, mais_vendido_loja4], axis=1)
print(f"\n🔝 {x} Produtos Mais Vendidos por Loja:\n")
display(mais_vendidos_juntas)

menos_vendido_loja1 = loja['Produto'].value_counts().tail(x).reset_index()
menos_vendido_loja1.columns = ['Produto', 'Qtd']
menos_vendido_loja2 = loja2['Produto'].value_counts().tail(x).reset_index()
menos_vendido_loja2.columns = ['Produto', 'Qtd']
menos_vendido_loja3 = loja3['Produto'].value_counts().tail(x).reset_index()
menos_vendido_loja3.columns = ['Produto', 'Qtd']
menos_vendido_loja4 = loja4['Produto'].value_counts().tail(x).reset_index()
menos_vendido_loja4.columns = ['Produto', 'Qtd']

# Asignar MultiIndex a las columnas para cada loja
menos_vendido_loja1.columns = pd.MultiIndex.from_product([['Loja 1'], menos_vendido_loja1.columns])
menos_vendido_loja2.columns = pd.MultiIndex.from_product([['Loja 2'], menos_vendido_loja2.columns])
menos_vendido_loja3.columns = pd.MultiIndex.from_product([['Loja 3'], menos_vendido_loja3.columns])
menos_vendido_loja4.columns = pd.MultiIndex.from_product([['Loja 4'], menos_vendido_loja4.columns])

# Concatenar lado a lado
menos_vendidos_juntas = pd.concat([menos_vendido_loja1, menos_vendido_loja2, menos_vendido_loja3, menos_vendido_loja4], axis=1)
print(f"\n🔻 {x} Produtos Menos Vendidos por Loja:\n")
display(menos_vendidos_juntas)
import pandas as pd
import matplotlib.pyplot as plt


# 4. Produtos Mais Vendidos (Gráfico de Pontos)

x = 4  # Número de produtos mais vendidos a serem exibidos

def plot_top_products(df, loja_num):
    mais_vendido = df['Produto'].value_counts().head(x).reset_index()
    mais_vendido.columns = ['Produto', 'Qtd']
    plt.figure(figsize=(6, 3))
    plt.scatter(mais_vendido['Produto'], mais_vendido['Qtd'], s=100, c='skyblue', edgecolors='black', label=f'Loja {loja_num}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.title(f'Top {x} Produtos Mais Vendidos - Loja {loja_num}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.legend()
    plt.show()

plot_top_products(loja, 1)
plot_top_products(loja2, 2)
plot_top_products(loja3, 3)
plot_top_products(loja4, 4)

def plot_down_products(df, loja_num):
    menos_vendido = df['Produto'].value_counts().head(x).reset_index()
    menos_vendido.columns = ['Produto', 'Qtd']
    plt.figure(figsize=(6, 4))
    plt.scatter(menos_vendido['Produto'], menos_vendido['Qtd'], s=100, c='skyblue', edgecolors='black', label=f'Loja {loja_num}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.title(f'Top {x} Produtos Menos Vendidos - Loja {loja_num}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.legend()
    plt.show()

plot_down_products(loja, 1)
plot_down_products(loja2, 2)
plot_down_products(loja3, 3)
plot_down_products(loja4, 4)
```

# 5. Frete Médio por Loja


```python
# prompt: faça um codigo para calcular o frete medio por loja e gere um grafico de área

# Calcula o frete médio para cada loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Cria um gráfico de barras para o frete médio
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

plt.figure(figsize=(10, 6))
plt.fill_between(lojas, fretes_medios, color='skyblue', alpha=0.4) # Usa fill_between para criar o gráfico de área
plt.plot(lojas, fretes_medios, marker='o', color='skyblue') # Adiciona uma linha para melhor visualização dos valores
plt.xlabel('Lojas')
plt.ylabel('Frete Médio')
plt.title('Frete Médio por Loja')
plt.show()

print(f"Frete Médio Loja 1: R${frete_medio_loja1:,.2f}")
print(f"Frete Médio Loja 2: R${frete_medio_loja2:,.2f}")
print(f"Frete Médio Loja 3: R${frete_medio_loja3:,.2f}")
print(f"Frete Médio Loja 4: R${frete_medio_loja4:,.2f}")
```
